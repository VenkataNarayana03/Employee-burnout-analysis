```python
if all(col in data.columns for col in ['Company Type','WFH Setup Available','Gender']):
    data=pd.get_dummies(data,columns=['Company Type','WFH Setup Available','Gender'],drop_first=True)
    data.head()
    encoded_columns=data.columns
else:
    print("one or more columns specified are not present")
    print(data.columns)
```

```python
y=data['Burn Rate']
x=data.drop(['Burn Rate'],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,shuffle=True,random_state=1)
scaler=StandardScaler()
scaler.fit(x_train)
x_train=pd.DataFrame(scaler.transform(x_train),index=x_train.index,columns=x_train.columns)
x_test=pd.DataFrame(scaler.transform(x_test),index=x_test.index,columns=x_test.columns)
```

```python
x_train
```

```python
y_train
```

```python
linear_regression_model=LinearRegression()
linear_regression_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import root_mean_squared_error

print("Linear Regression model performnace metrics : \n")
y_predict=linear_regression_model.predict(x_test)

error1=mean_squared_error(y_test,y_predict)
print("mean squared error       :",error1)

error2=root_mean_squared_error(y_test,y_predict)
print("root mean squared error  :",error2)

error3=r2_score(y_test,y_predict)
print("r2 score                 :",error3)
```
